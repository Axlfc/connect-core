## Setup

```python
import sys
import os
sys.path.insert(0, os.path.abspath('.'))

# Import agent modules
from agents.agent_router import AgentRouter
from agents.multi_agent_orchestrator import MultiAgentOrchestrator
from agents.task_decomposer import TaskDecomposer

print("✓ Imports successful")
```

## Test 1: Agent Router

```python
# Initialize agent router
router = AgentRouter()

# Test task routing
test_tasks = [
    "Analyze sentiment of customer reviews",
    "Generate images for product catalog",
    "Extract information from documents",
    "Process data pipeline",
    "Create summary report"
]

print("Testing Agent Router:")
print("-" * 50)

for task in test_tasks:
    try:
        agent = router.route(task)
        print(f"✓ Task: {task}")
        print(f"  → Routed to: {agent}")
        print()
    except Exception as e:
        print(f"✗ Error routing '{task}': {e}")
        print()
```

## Test 2: Multi-Agent Orchestration

```python
# Initialize orchestrator
orchestrator = MultiAgentOrchestrator()

# Test orchestration
print("Testing Multi-Agent Orchestration:")
print("-" * 50)

test_workflow = {
    "workflow_id": "test_wf_001",
    "tasks": [
        {
            "id": "task_1",
            "description": "Extract data from source",
            "type": "extraction"
        },
        {
            "id": "task_2",
            "description": "Process and enrich data",
            "type": "processing",
            "depends_on": ["task_1"]
        },
        {
            "id": "task_3",
            "description": "Generate insights",
            "type": "analysis",
            "depends_on": ["task_2"]
        }
    ]
}

try:
    result = orchestrator.orchestrate(test_workflow)
    print(f"✓ Orchestration successful")
    print(f"  Result: {result}")
except Exception as e:
    print(f"✗ Orchestration error: {e}")
```

## Test 3: Task Decomposition

```python
# Initialize task decomposer
decomposer = TaskDecomposer()

# Test task decomposition
complex_task = "Analyze customer sentiment from reviews, generate summary report, and create visualizations"

print("Testing Task Decomposition:")
print("-" * 50)
print(f"Complex Task: {complex_task}")
print()

try:
    subtasks = decomposer.decompose(complex_task)
    print(f"✓ Task decomposed into {len(subtasks)} subtasks:")
    print()
    for i, subtask in enumerate(subtasks, 1):
        print(f"  {i}. {subtask}")
except Exception as e:
    print(f"✗ Decomposition error: {e}")
```
