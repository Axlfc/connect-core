## Setup

```python
import requests
import json
import os
from datetime import datetime

# Configuration
N8N_BASE_URL = os.getenv('N8N_BASE_URL', 'http://localhost:5678')
OLLAMA_BASE_URL = os.getenv('OLLAMA_BASE_URL', 'http://localhost:11434')
QDRANT_BASE_URL = os.getenv('QDRANT_BASE_URL', 'http://localhost:6333')

print(f"N8N: {N8N_BASE_URL}")
print(f"Ollama: {OLLAMA_BASE_URL}")
print(f"Qdrant: {QDRANT_BASE_URL}")
```

## Test 1: Check Service Health

```python
print("Checking service health...")
print("-" * 50)

# Check n8n health
try:
    response = requests.get(f"{N8N_BASE_URL}/healthz", timeout=5)
    print(f"✓ n8n: {response.status_code} - {response.text[:50]}")
except Exception as e:
    print(f"✗ n8n: {str(e)[:50]}")

# Check Ollama health
try:
    response = requests.get(f"{OLLAMA_BASE_URL}/api/tags", timeout=5)
    print(f"✓ Ollama: {response.status_code}")
except Exception as e:
    print(f"✗ Ollama: {str(e)[:50]}")

# Check Qdrant health
try:
    response = requests.get(f"{QDRANT_BASE_URL}/health", timeout=5)
    print(f"✓ Qdrant: {response.status_code}")
except Exception as e:
    print(f"✗ Qdrant: {str(e)[:50]}")
```

## Test 2: Ollama Integration

```python
print("Testing Ollama Integration:")
print("-" * 50)

# Get available models
try:
    response = requests.get(f"{OLLAMA_BASE_URL}/api/tags")
    models = response.json()
    print(f"✓ Available models: {len(models.get('models', []))}")
    for model in models.get('models', []):
        print(f"  - {model['name']}")
except Exception as e:
    print(f"✗ Error fetching models: {e}")

print()
```

## Test 3: Generate Text with Ollama

```python
print("Testing Text Generation with Ollama:")
print("-" * 50)

test_prompt = "What are the main benefits of containerization in AI systems? Answer in 2 sentences."
model = "llama3.2"  # You can change this to another available model

try:
    payload = {
        "model": model,
        "prompt": test_prompt,
        "stream": False
    }
    
    response = requests.post(f"{OLLAMA_BASE_URL}/api/generate", json=payload, timeout=30)
    
    if response.status_code == 200:
        result = response.json()
        print(f"✓ Model: {model}")
        print(f"\nPrompt: {test_prompt}")
        print(f"\nResponse: {result['response']}")
    else:
        print(f"✗ Error: {response.status_code}")
        print(response.text)
except Exception as e:
    print(f"✗ Error: {e}")
```

## Test 4: Qdrant Vector Database

```python
print("Testing Qdrant Vector Database:")
print("-" * 50)

# Create a test collection
collection_name = "test_documents"

try:
    # Create collection
    collection_config = {
        "vectors": {
            "size": 384,
            "distance": "Cosine"
        }
    }
    
    response = requests.put(
        f"{QDRANT_BASE_URL}/collections/{collection_name}",
        json=collection_config,
        timeout=5
    )
    
    if response.status_code in [200, 201]:
        print(f"✓ Collection '{collection_name}' created/exists")
    else:
        print(f"✓ Collection status: {response.status_code}")
except Exception as e:
    print(f"✗ Error: {e}")
```
